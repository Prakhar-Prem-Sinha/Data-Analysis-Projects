
#Handle the missing value columns with the proper assumtion.

```python
iris.isnull()
```

```python
iris.isnull().sum()
```

```python
iris.dropna(axis=1)
```

```python
iris
```

```python
# now replace the missing values of  iris by its mean value using fillna() method
```

```python
saple_length=iris['sepal_length']
saple_length
```

```python
saple_length.fillna(saple_length.mean(), inplace=True)
```

```python
saple_width=iris['sepal_width']
saple_width
```

```python
saple_width.fillna(saple_width.mean(), inplace=True)
```

```python
petal_width=iris['petal_width']
petal_width
```

```python
petal_width.fillna(petal_width.mean(), inplace=True)
```

```python
petal_length=iris['petal_length']
petal_length
```

```python
petal_length.fillna(petal_length.mean(), inplace=True)
```
